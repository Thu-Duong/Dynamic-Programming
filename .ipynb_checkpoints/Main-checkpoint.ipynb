```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#from scipy.interpolate import interp1d
#import scipy
#from numba import njit, prange
#from function import *
#from looping import *
#from quantecon.optimize.scalar_maximization import brent_max
```

```python
beta = 0.95
sigma = 2.0
alpha = 0.3
delta = 0.1
max_iter = 500
crit = 1e-6
```

```python
def utility_crra(argument, exponent):
    """Return the CRRA utility evaluation of *argument* where the Arrow-Pratt
    coefficient of relative risk aversion takes on the value *exponent*.
    Neither handle cases where *argument* is less than nor log utility.
    Utility = (C^(1 - sigma) - 1)/(1 - sigma)
    """

    assert argument >= 0, "Only positive quantities allowed for *argument*."
    assert exponent != 1, "No provision made for log utility."

    numerator = argument ** (1.0 - exponent)
    denominator = 1.0 - exponent
    out = (numerator - 1) /denominator
    return out
```

```python
def steady_state_crra(beta, delta, alpha, sigma):
    """
    
    """
    capitals = ( (1-beta+delta*beta) / (alpha*beta) )**(1/(alpha-1))
    consumptions = capitals**alpha - delta*capitals

    coeff = -(1.0/sigma)*consumptions*alpha*(alpha-1.0)*capitals**(alpha - 2.0);
    A = np.zeros( (2, 2) )
    A[0][0] = 1.0 + beta*coeff
    A[0][1] = -coeff
    A[1][0] = -1
    A[1][1] = 1.0/beta

    return capitals, consumptions, A
```

```python
def bellman_equation(u, val_old, beta):
    """
        The Bellman operator, which computes and returns the updated
        value function `Tv` for a value function `v`.

        Parameters
        ----------
        val_old : array_like(float, ndim=1)
            Value function vector, of length n.

        u : array_like( 2-dimensional ndarray of shape (n, m))
            value grid.

        Returns
        -------
        Tv : ndarray(float, ndim=1)
            Updated value function vector, of length n.

        """

    return u + beta*np.transpose(val_old)
```

```python
capitals, consumptions, _ = steady_state_crra(beta, delta, alpha, sigma)
grid_max = 1.2*capitals
grid_min = 0.8*capitals 
num_states = 400
```

```python
def state_wise_max(v):
    """
    Given utility grid with certain reward
    
    Parameters
        ----------
        v : array_like( 2-dimensional ndarray of shape (n, m))
            value grid

        Returns
        -------
        value_fn : array_like(float, ndim=1)
            Value function vector, of length n.

        pol_index : ndarray(int, ndim=2)
            policy index, of shape (n, n).
    """
    value_fn = np.max(v, axis = 0)
    pol_index = np.argmax(v, axis = 0)
    return value_fn, pol_index
```

```python
def evaluate_policy(policy, u_grid, beta, num_states):
    """
    Compute the value of a policy.

    Parameters
        ----------
        policy : array_like(int, ndim=1)
            Policy vector, of length n.
            array([2, 0, 2, 1], dtype=int64)


    """
    # assert beta == 1, "provide beta between 0 and 1"

    # Solve (I - beta * Q_policy) v = U_policy
    U_policy = u_grid[range(len(policy)), policy]
    Q_policy = np.zeros((num_states, num_states))
    Q_policy[policy] = 1
    
    _I = np.identity(num_states)
    A = _I - beta * Q_policy

    v_policy = np.linalg.solve(A, U_policy)
    return v_policy
```

```python
def v_greedy(v, u_grid, beta):
    """
    Parameters
        ----------
        v : array_like(float, ndim=1)
            Value function vector, of length n.

        policy : ndarray(int, ndim=1), optional(default=None)
            Optional output array for `sigma`.
    Returns
        -------
        policy : ndarray(int, ndim=1)
            v-greedy policy vector, of length n.
    """
    #if policy is None:
    #    policy = np.empty(num_states, dtype=int)
    
    #improve value function
    new_value = bellman_equation(u_grid, v, beta)
    
    #find the new_policy
    _ , new_policy = state_wise_max(new_value)
    return new_policy
```

```python
k_grid = np.linspace(grid_min, grid_max, num_states)
u_grid =  np.zeros((num_states, num_states))
```

```python
#looping over u
for j, k_state in enumerate(k_grid):
    for i, k_action in enumerate(k_grid):
        c_choice = k_state**alpha + (1 - delta)*k_state - k_action
        if c_choice > 0:
            u_grid[i,j] = utility_crra(c_choice, sigma)
```

### value iteration

```python
# value iteration
val_store_iter = np.empty((max_iter,num_states))
val_store_iter[:] = np.NaN
val_old = np.zeros(num_states)
val_iter = np.zeros(num_states)

for i in range(max_iter):
    value_function = bellman_equation(u_grid, val_old, beta)
    val_iter = np.max(value_function, axis = 0)
    k_policy = np.argmax(value_function, axis = 0)
        
    max_diff = np.max(np.absolute(val_iter - val_old))
    val_store_iter[i,:] = val_iter
    # quit iterations, when convergence is achieved
    if max_diff < crit:
        break
    val_old[:] = val_iter
```

```python
num_iter = i
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
x = k_grid
for j in range(num_iter):
    y = val_store_iter[j,:]
    ax.plot(x, y, linewidth=2, alpha=0.6)
#ax.legend()
plt.show()
```

### policy iteration

```python
#first setsup
v_init, _ = state_wise_max(u_grid)
policy = v_greedy(v_init, u_grid, beta)

new_policy = np.empty(shape=[num_states, 0], dtype=int)

for i in range(max_iter):
    # Policy evaluation
    v_policy = evaluate_policy(policy, u_grid, beta, num_states)
    # Policy improvement
    new_policy = v_greedy(v_policy, num_states, u_grid, beta)
    
    if np.array_equal(new_policy, policy):
        break
    policy[:] = new_policy

num_iter = i + 1
```
