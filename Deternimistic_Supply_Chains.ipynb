## Introduction - Inventory Policies

We analyzed three policies:
(s, Q) A continuous review policy with a reorder point s and a fixed order quantity Q.
+ Pros: low amount of safety stock needed; optimized order quantity.
- Cons: continuous review needed; multiple items cannot be grouped in one
order with one supplier.

(R, S) A periodic review policy following a review period R and an order up-to level S.
+ Pros: multiple items can be grouped in one order with one supplier.

quantity.
